```python
import os
import esprima
import pandas as pd
import re
import json
from nltk.tokenize import word_tokenize
import sys
import random
```

```python
def list_files(dir):                                                                                                  
    r = []                                                                                                            
    subdirs = [x[0] for x in os.walk(dir)]
    print("subdirs ", subdirs)
    for subdir in subdirs:                                                                                            
        files = os.walk(subdir).__next__()[2]                                                                             
        if (len(files) > 0):                                                                                          
            for file in files:                                                                                        
                r.append(os.path.join(subdir, file))                                                                         
    return r  
```

```python
files = list_files("./Data/filtered_sites")
```

```python
def tokenize(answer):
    try:
        esprima.parseScript(answer)
        tokenized = esprima.tokenize(answer)
        result = []
        for token in tokenized:
            result.append(token.value)
        print("no error")
        return result
    except:
        print("error")
        return None
```

```python
def buildInstance(code, i):
    current_line = {}
    current_line['repo'] = ""
    current_line['path'] = ""
    current_line['function'] = ""

    url_idx = file.find('filtered_sites')
    current_line['url'] = file[url_idx+15:]+"_"+str(i)


    current_line['sha'] = ""
    current_line['partition'] = ""
    current_line['language'] = "javascript"
    current_line['original_string'] = code
    current_line['code'] = code
    current_line['code_tokens'] = tokenize(code)
    if (not(current_line['code_tokens']) or len(current_line['code_tokens'])< 4 or len(current_line['code_tokens']) > 256):
        print("None")
        return None
    current_line['docstring'] = ""
    current_line['docstring_tokens']  = ""

    return current_line    
```

```python
all_dict = []
count = 0
for file in files:
    print(count)
    count+=1
    text_file = open(file, "r")

    #read whole file to a string
    data = text_file.read()

    #close file
    text_file.close()
    
    code_list = data.split(";")
    
    for i in range(len(code_list)):
        current_dict = buildInstance(code_list[i], i)
        if current_dict:
            all_dict.append(current_dict)

print(len(all_dict))
```
