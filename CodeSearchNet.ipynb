```python
data = pd.read_pickle("original_word_df")
```

```python
# data.loc[data['code'] != None]
data = data[~data['code'].isnull()]
```

```python
data.drop_duplicates(inplace=True)
```

```python
def tokenize_code(answer):
    try:
        print(answer)
#         esprima.parseScript(answer)
        tokenized = esprima.tokenize(answer)
        
        result = "["
        for token in tokenized:
            result += token.value
            result += ","
        result = result[:-1]
        result += "]"
        return result
    except:
        print("Unexpected error:", sys.exc_info()[0])
        return None
```

```python
program = "# my_form.cgi use strict; use warnings; use utf8; use HTTP::Headers; use Template;  my $headers = HTTP::Headers->new; $headers->header('Content-Type' =>  'text/html;charset=UTF-8'); print $headers->as_string;my %urls = ( 'https://www.google.com'    => 'google' , 'https://www.msn.com'       => 'msn', 'https://www.yahoo.com'     => 'yahoo'  ); my $template = Template->new; $template->process('my_form.ttml', { urls => \%urls });"
tokenize_code(program)
```

```python
data[data['code_tokens'].isnull()]
```

```python
question_ids = data['docstring'].unique()
```

```python
for question in question_ids:
    code = data['code'][data['docstring'] == question].unique()[0]
    code_tokens = tokenize_code(code)
    data['code_tokens'][data['docstring'] == question] = code_tokens
data = data[~data['code_tokens'].isnull()]
```

```python
data
```

```python
def clean_token(tokenized_words):
    # in the form of [hsjsjs, hsdjjsd]
    print(tokenized_words)
    tokenized_words = tokenized_words[1:-1]
    tokenized_words = tokenized_words.split(",")
    
    return tokenized_words
    
```

```python
data_dict = data.to_dict(orient='records') #list of dictionaries
for single in data_dict:
    single["code_tokens"] = clean_token(single["code_tokens"])
    single["docstring_tokens"] = clean_token(single["docstring_tokens"])
```
