```python
files = ["javascript1067_1300", "javascript651_1066", "javascript1951_2600_with_code", "javascript1301_1869_with_code", "javascript1951_2600_with_answers"]

# data = pd.read_pickle("./Data/scraped_javascript_questions/javascript1301_1869_with_code")
titles = set()
for file in files:
    data = pd.read_pickle("./Data/scraped_javascript_questions/"+file)
    titles |= set(data['title'])
```

```python
titles = str(titles).lower()

titles = remove_stopwords(titles)
```

```python
tokens = nltk.word_tokenize(titles)

# titles = remove_stopwords(titles)
```

```python
tagged = nltk.pos_tag(tokens)

```

```python
tagged[6:100]
```

```python
words = []
for i in range(len(tagged)):
    if tagged[i][1] not in ["VBG", "NN", "JJ", "NNP", "CC", "VPD", "VB", ".", "''", "VBD", "RB", ":", ",", ]:
        words.append(tagged[i][0])
   
        
```

```python
words
```

```python
titles_split = titles.split()
  
# Pass the split_it list to instance of Counter class.
Counter = Counter(titles_split)
  
# most_common() produces k frequently encountered
# input values and their respective counts.
most_occur = Counter.most_common(30)
  
print(most_occur)
```
