```python
import pandas as pd
import json
```

```python
#get all the labels
f = open("./Data/NoThirdParty/no_third_party.output", "r")
labels = []
while(f.readline()):
    line = f.readline()
    idx = line.find("Returns")
    labels.append(line[idx+8:-1])

```

```python
tags = ["menu", "header", "social-media", "footer", "navbar", "banner", "nav-bar", "navicon", "content"]
```

```python
#get all the labels
f2 = open("./Data/NoThirdParty/test.jsonl", "r")
allCode = []
count = 0
while(f2.readline()):
    line = f2.readline()
    print(count)
    count+=1
    try:
        code = json.loads(line)['code']
        
        url = json.loads(line)['url'].replace("/", "-")
        js_idx = url.find(".js")
        url = url[:js_idx] + "-function" + url[js_idx+4:]
        
    except:
        print("not found")
        code = ""
        url = ""
    allCode.append({'num': count,'code': code, 'website': url})
    

```
