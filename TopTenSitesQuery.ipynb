```python
import pandas as pd
import json
```

```python
import pymysql.cursors
from pprint import pprint
```

```python
dict_list = []
REMOTE_IP = "comnetsad.dyndns.org"

connection = pymysql.connect(host=REMOTE_IP,
                                 user='russell',
                                 port=9922,
                                 password='RusNYUAD@2021',
                                 db='clonedSites',
                                 autocommit=True)

sites = ["https://mitpress.mit.edu/", "https://twin-cities.umn.edu/", "https://uci.edu/",
 "https://umich.edu/", "https://www.berkeley.edu/", "https://www.cmu.edu/", "https://www.harvard.edu/", "https://www.uchicago.edu/", "https://www.utexas.edu/", "https://www.washington.edu/"]

with connection.cursor(pymysql.cursors.DictCursor) as cursor:
    for site in sites:
        query = f'SELECT * FROM cachedPages WHERE initiatingUrl="{site}" AND type LIKE "%javascript%"'
        cursor.execute(query)

        if cursor.rowcount > 0:
            results = cursor.fetchall()
            for r in results:
                dict_list.append(r)

            pprint(results)
```
